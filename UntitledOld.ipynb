# 4. CREATE A LINEAR REGRESSION MODEL AND DISCUSS THE MODEL PARAMETERS

```r
#Linear Model
pl <- file$petal.length
pw <- file$petal.width

lmodal <- lm(pl ~ pw)
summary(lmodal)
```

```r
plot(lmodal)
```

```r
scatter.smooth(pl , pw , main = "petal.Length v/s petal.Width")
```

# It can be seen from the above plot that the data points are increasing along with the Linear Modal Line thus having a increasing linear relationship with 'petal.length' and 'petal.width'
