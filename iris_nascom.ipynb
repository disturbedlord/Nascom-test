# 4. CREATE A LINEAR REGRESSION MODEL AND DISCUSS THE MODEL PARAMETERS

```r
#Linear Model
sp <- file$species
pw <- file$petal_width

lmodal <- lm(sp ~ pw)
```

```r
plot(lmodal)
```

```r
scatter.smooth(sp , pw ,xlab="Petal Width", ylab="Species", main="Species plotted v/s Petal Width")
```

# It can be seen from the above plot that the data seems to grow along the linear modal but is not spread throughout the linear modal line and thus is not a proper linear Modal.
